# feature_selection

## 1. VarianceThreshold

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import VarianceThreshold

import matplotlib.pyplot as plt 
import numpy as np
```

```python
iris = load_iris()

X = iris.data 
y = iris.target 

plt.plot(X)
plt.legend(iris.feature_names)
```

```python
X.var(axis=0)
```

```python
selector = VarianceThreshold(threshold=0.2)
selector.fit_transform(X=X)
```

## selector.get_support(), selector.get_params

```python
selector.get_params(), selector.get_support()
```

### Boolean indexing to get selected features

```python
np.array(iris.feature_names)[selector.get_support()]
```

# selectKBest

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
selectorKbest = SelectKBest(chi2, k=2)
```

```python
selectorKbest.fit_transform(X, y), selectorKbest.get_support()
```
