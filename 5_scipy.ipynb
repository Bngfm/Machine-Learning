# Optimization

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(0, 2, 100)
y = 1/3*x**3 - 3/5 * x**2 +2+ np.random.randn(x.shape[0])/20

plt.scatter(x, y)
```

## model 

```python
def f(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
from scipy import optimize
```

```python
params , params_covariance = optimize.curve_fit(f, x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]))
```

# minimisation

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import optimize
```

```python
def funct(x):
    return x**2 + 15*np.sin(x)
```

```python
x = np.linspace(-15, 15, 200)
plt.plot(x, funct(x))
```

```python
optimize.minimize(funct, x0=-10)
```

```python
min_X = optimize.minimize(funct, x0=-10).x
min_X
```
