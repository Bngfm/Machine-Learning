```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

## Linear Regression

```python
np.random.seed(0)
m = 100 # creating m samples
X = np.linspace(0, 10, m).reshape(m, 1)
y = X + np.random.randn(m, 1)
```

```python
plt.scatter(X, y) 
```

## Model

```python
model = LinearRegression()
```

## Train 

```python
model.fit(X, y)
```

## Score 

```python
model.score(X, y)
```

## predictions

```python
predictions = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, predictions)
```

# Classification

```python
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier
```

```python
titanic = pd.read_excel('titanic3.xls')
titanic.shape, titanic.head(3)
```

```python
titanic = titanic[['survived', 'pclass', 'sex', 'age']]
titanic.dropna(axis=0, inplace=True)
titanic['sex'].replace(['male', 'female'], [0, 1], inplace=True)

titanic.head(2)
```

### get Y (target) | get X (features)

```python
y = titanic['survived']
X = titanic.drop('survived', axis=1)
```

 ## model 

```python
model2 = KNeighborsClassifier(n_neighbors=1)
```

## fit 

```python
model2.fit(X, y)
```

## score 

```python
model2.score(X, y)
```

## predict 

```python
model2.predict(X)
```

```python
def survived(model, pclass=0, sex=0, age=10):
    X = np.array([pclass, sex, age])
    X = X.reshape(1, 3)
    print(model.predict(X))
    print(model.predict_proba(X))
```

```python
survived(model2, pclass=2, sex=1, age=25)
```
