# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(1, 5, 10)
Y = X ** 2 
plt.plot(X, Y)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X, Y)
plt.plot(X, X**3, label="first")
plt.plot(X, X**4, label="second")
plt.title("Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

plt.savefig("figure.png")
```

## subplot 

```python
plt.subplot(1,2,1)
plt.plot(X, Y)
plt.show()

plt.subplot(1, 2, 2)
plt.scatter(X, X**2)
plt.show()

plt.subplot(2, 2, 3)
plt.plot(X, Y**3)
plt.show()

plt.subplot(2, 2, 4)
plt.plot(X, Y**4)
plt.show()
```
